```python
# @title Social Media Username Checker (Colab Version)
# @markdown Run all cells (Runtime ‚Üí Run all) then enter a username

import requests
from IPython.display import HTML, display
import time

def check_username(username):
    platforms = [
        # Major Platforms
        {"name": "Twitter", "url": f"https://twitter.com/{username}", "icon": "https://abs.twimg.com/favicons/twitter.ico"},
        {"name": "Instagram", "url": f"https://instagram.com/{username}", "icon": "https://www.instagram.com/static/images/ico/favicon.ico"},
        {"name": "Facebook", "url": f"https://facebook.com/{username}", "icon": "https://static.xx.fbcdn.net/rsrc.php/yD/r/d4ZIVX-5C-b.ico"},
        {"name": "TikTok", "url": f"https://tiktok.com/@{username}", "icon": "https://s16.tiktokcdn.com/musical/resource/wap/static/image/logo_144c91a.png"},

        # Tech & Coding
        {"name": "GitHub", "url": f"https://github.com/{username}", "icon": "https://github.githubassets.com/favicons/favicon.ico"},
        {"name": "GitLab", "url": f"https://gitlab.com/{username}", "icon": "https://gitlab.com/favicon.ico"},

        # Video Platforms
        {"name": "YouTube", "url": f"https://youtube.com/@{username}", "icon": "https://www.youtube.com/s/desktop/d5a0b8c0/img/favicon.ico"},
        {"name": "Twitch", "url": f"https://twitch.tv/{username}", "icon": "https://static.twitchcdn.net/assets/favicon-32-e29e246c.png"},

        # Professional
        {"name": "LinkedIn", "url": f"https://linkedin.com/in/{username}", "icon": "https://static.licdn.com/sc/h/akt4ae504epesldzj74dzred8"},

        # Add more platforms as needed...
    ]

    results = []

    for platform in platforms:
        try:
            start_time = time.time()
            response = requests.get(platform["url"], headers={"User-Agent": "Mozilla/5.0"}, timeout=5)
            elapsed_time = round(time.time() - start_time, 2)

            status = "üü¢ AVAILABLE" if response.status_code == 404 else "üî¥ TAKEN"
            if "login" in response.url or "signup" in response.url:
                status = "üü° UNKNOWN (Check Manually)"

            results.append(f"""
            <tr>
                <td><img src="{platform['icon']}" width="20" style="vertical-align:middle"> {platform['name']}</td>
                <td>{status}</td>
                <td>{elapsed_time}s</td>
                <td><a href="{platform['url']}" target="_blank">Visit</a></td>
            </tr>
            """)
        except Exception as e:
            results.append(f"""
            <tr>
                <td><img src="{platform['icon']}" width="20" style="vertical-align:middle"> {platform['name']}</td>
                <td>‚ö†Ô∏è ERROR</td>
                <td>N/A</td>
                <td><a href="{platform['url']}" target="_blank">Try</a></td>
            </tr>
            """)

    display(HTML(f"""
    <h2>Results for @{username}</h2>
    <table border="1" cellpadding="8" style="border-collapse: collapse; width: 100%">
        <tr>
            <th>Platform</th>
            <th>Status</th>
            <th>Response Time</th>
            <th>Link</th>
        </tr>
        {''.join(results)}
    </table>
    <p>Note: Some platforms may give false results due to rate limiting</p>
    """))

# @markdown ---
# @markdown ### Enter username below and run this cell:
username = "elonmusk" # @param {type:"string"}
check_username(username.strip("@"))
```
